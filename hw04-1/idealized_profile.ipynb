```python
import numpy as np
import matplotlib.pyplot as plt

nz=101

th=np.zeros([nz])

z=0.1*np.arange(th.shape[0])+0.05
H=0.05
zcr=1.
inv_str=10.
env_sta=10.

th=env_sta*(1.-np.exp((2.5-z))/(1.+np.exp((2.5-z))))+300.
th=th+inv_str*np.exp((z-zcr)/H)/(1.+np.exp((z-zcr)/H))


H=0.05
zcr=1.
inv_str=-0.2
env_sta=0.4

rh=env_sta*np.exp((2.5-z))/(1.+np.exp((2.5-z)))+(1-env_sta)
rh=rh+inv_str*np.exp((z-zcr)/H)/(1.+np.exp((z-zcr)/H))

#print(z)
#print(th)

fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(2,2),dpi=300)
ax.plot(rh,z)
ax.set_ylim([0,5])
ax.set_xlim([0,1])
```
